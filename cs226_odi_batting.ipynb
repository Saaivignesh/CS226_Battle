```python
cur.execute("SELECT player, hundreds FROM odi_data WHERE hundreds >= 10 ORDER BY hundreds DESC;")
rows = cur.fetchall()
print("\n-- Query 1a: Players with >=10 centuries --")
display(pd.DataFrame(rows, columns=["player","hundreds"]))
```

```python
cur.execute("SELECT player FROM odi_data WHERE player LIKE 'S%';")
rows = cur.fetchall()
print("\n-- Query 1e: Players starting with 'S' --")
display(pd.DataFrame(rows, columns=["player"]))
```

```python
cur.execute("SELECT player, runs FROM odi_data WHERE runs > 12000 ORDER BY runs DESC;")
rows = cur.fetchall()
print("\n-- Query 2a: Players with >12,000 runs --")
display(pd.DataFrame(rows, columns=["player","runs"]))
```

```python
cur.execute("""
SELECT player, span
FROM odi_data
WHERE CAST(split_part(span, '-', 1) AS INTEGER) > 2000;
""")
rows = cur.fetchall()
print("\n-- Query 2b: Players who debuted after 2000 --")
display(pd.DataFrame(rows, columns=["player","span"]))
```
